## Plotting a sine wave using Matplotlib and Numpy!


````python
import numpy as np
import matplotlib.pyplot as plt


x = np.arange(0, 6.28, 0.01)   # step =0.1,start index =0,end index =6.28.

# Sine function
y = np.sin(x)

# Plot
plt.plot(x, y,linewidth=1.5)   
plt.title("Sine Wave")
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.grid(True)

plt.show()

````
